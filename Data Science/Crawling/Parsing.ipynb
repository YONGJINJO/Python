```python
response = requests.get('https://www.walmart.com/search/?query=monitor')
soup = BeautifulSoup(response.text, 'html.parser')
soup
```

```python
last_li_tag = soup.select('.paginator-list li')[-1]
page_length = last_li_tag.a.text
```

```python
page=[]
for i in range(1, int(page_length) + 1):
    response = requests.get('https://www.walmart.com/search/?page=%d&query=monitor'.format(i))
    page.append(response.text)
    
page
```

```python
import time
import requests
from bs4 import BeautifulSoup

# 빈 리스트 생성
pages = []
page_num = 1

while True:
    response = requests.get('http://www.ssg.com/search.ssg?target=all&query=nintendo&page=' + str(page_num))
    soup = BeautifulSoup(response.text, 'html.parser')
    if len(soup.select('.csrch_tips')) == 0:
        pages.append(soup)
        print(str(page_num) + "번째 페이지 가져오기 완료")
        page_num += 1
        time.sleep(3)
    else:
        break
print(len(pages))
```

```python
response = requests.get('https://workey.codeit.kr/ratings/index')
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
rating_pages = []
for year in range(2010, 2019):
    for month in range(1, 13):
        index = 0
        while True:
            response = requests.get('https://workey.codeit.kr/ratings/index?year=' + str(year) + '&month=' + str(month) + '&weekIndex=' + str(index))
            soup = BeautifulSoup(response.text, 'html.parser')
            li = soup.select('.row')
            if len(li) > 1:       
                rating_pages.append(response.text)
                index += 1
            else:
                break
            
len(rating_pages)
```

## 데이터의 이상점을 판단하는 기준?

### IQR  : 75%지점과 25% 지점간 거리  
### 75% 위로 혹은 25% 아래로 1.5 IQR 이상 떨어진 거리의 DATA는 이상점이라 한다

## 이상점이 제대로 된 데이터라면?  
### 분석에 방해가 되면 제거하고 의미있는 정보라면 그대로 둔다

```python
df = pd.read_csv('C:/Users/choyj/Desktop/Python/Data_Science/Fine_Data/beer.csv', index_col = 0)
df
```

```python
df.plot(kind = 'box', y = 'abv')
```

```python
df['abv'].describe()
```

```python
q1 = df['abv'].quantile(0.25)
```

```python
q3 = df['abv'].quantile(0.75)
```

```python
iqr = q3 - q1
```

```python
condition = (df['abv'] < q1 - 1.5 * iqr) | (df['abv'] > q3 + 1.5 * iqr)
```

```python
df[condition]
```

```python
df.loc[2250, 'abv'] = 0.055
```

```python
df.loc[2250]
```

```python
df[condition]
```

```python
df[condition].index
```

```python
df.drop(df[condition].index, inplace = True)
```

```python
df[condition]
```

```python
df.plot(kind = 'box', y = 'abv')
```

```python
df = pd.read_csv('C:/Users/choyj/Desktop/Python/Data_Science/DataFrame/enrolment_3.csv')
df
```

```python
room_types = df['room assignment'].value_counts()
room_types = list(room_types.index)
room_types.sort()
print(room_types)        
```

```python
for j in range(len(room_types) - 1):
    room_type = room_types[j]
    courses = df[df['room assignment'] == room_type]
    courses = courses['course name'].value_counts()
    courses = list(courses.index)
    courses.sort()
    i = 1
    room = room_type.split()[0]
    for course in courses:
        df.loc[(df['course name'] == course) & (df['room assignment'] == room_type), 'room assignment'] = '{}-{}'.format(room, i)
        i += 1
df
```
